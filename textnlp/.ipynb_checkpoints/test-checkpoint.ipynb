```python
import gutenberg.load_text as load_text
import pretokenize

import os
```

```python
from __init__ import default_ascii_dir, default_truncated_dir
```

```python
fbs = os.listdir(default_ascii_dir)
```

```python
fns = [os.path.join(default_ascii_dir, x) for x in fbs]
```

```python
fn = fns[2]
print(fn)
```

```python
txt = load_text.load_text(fn, write=True)
```

```python
new_txt, counts = pretokenize.main(txt, return_counts=True)
```

```python
import matplotlib.pyplot as plt
import wordcloud

try:
    counts.pop('.')
except:
    pass
try:
    counts.pop('p')
except:
    pass
wc_size = (14, 14)
sm = sum(counts.values())
frequency_dict = {k: v/sm for k, v in counts.items()}

wc = (
    wordcloud.WordCloud(
        width=1000, height=1000, 
        background_color='white',
        min_font_size=10
    ).generate_from_frequencies(frequency_dict)
)
fig = plt.figure(figsize=wc_size, facecolor=None)
ax = fig.add_subplot()
ax.set_title("Words in the disaster tweets")
_ = ax.imshow(wc) 
```
